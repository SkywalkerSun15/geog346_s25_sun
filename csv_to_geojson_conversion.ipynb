```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import shape
import json
import os

# Read the CSV file
df = pd.read_csv("data/Parks_Properties_20250402.csv")
df.columns = df.columns.str.strip()  # clean up column names

def parse_geometry(x):
    if isinstance(x, str) and x.strip():
        try:
            return shape(json.loads(x))
        except json.JSONDecodeError:
            return None
    return None

# Convert the 'multipolygon' column (GeoJSON string) to Shapely geometries
df['geometry'] = df['multipolygon'].apply(parse_geometry)

# Create a GeoDataFrame (assuming the coordinates are in WGS84)
gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

# Optional: inspect the first few rows
print(gdf.head())
```

```python
output_file = os.path.join("data", "Parks_Properties_20250402.shp")
gdf.to_file(output_file)
print(f"Shapefile has been written to {output_file}")
```
